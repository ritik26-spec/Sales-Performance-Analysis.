**4. Perform EDA to analyze total sales, average sales, and category performance**

```python
total_revenue = df['Revenue'].sum()
total_revenue
```

```python
total_profit = df['Profit'].sum()
total_profit
```

```python
avg_sales = df['Revenue'].mean()
avg_sales
```

```python
category_perf = df.groupby('Category').agg({'Revenue': 'sum', 
'Profit': 'sum', 'Quantity': 'sum', 'Order_ID': 'count'}).rename(columns={'Order_ID': 'Transaction_Count'}).sort_values(by='Revenue', ascending=False)
category_perf
```

```python
#Monthly Sales Trend
df['Month_Year'] = df['Order_Date'].dt.to_period('M')
monthly_sales = df.groupby('Month_Year')['Revenue'].sum().reset_index()
monthly_sales['Month_Year'] = monthly_sales['Month_Year'].astype(str)
monthly_sales = monthly_sales.sort_values(by='Month_Year')
```

**5. Visualize insights using appropriate plots**

```python
#Line Plot: Sales Trend Over Time
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(monthly_sales['Month_Year'], monthly_sales['Revenue'], marker='o', color='blue', linestyle='-', linewidth=2)
ax.set_title('Monthly Sales Revenue Trend', fontsize=14, fontweight='bold')
ax.set_xlabel('Month(Year-Month)', fontsize=12)
ax.set_ylabel('Total Revenue ($)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig('monthly_sales_trend.png')
plt.show()
```

```python
#Bar-Chart: Category-Wise Performance
category_perf = category_perf.reset_index()
fig, ax = plt.subplots(figsize=(8,6))
sns.barplot(data=category_perf,x='Category',y='Revenue', ax=ax, palette='Blues_r')
ax.set_title('Total Revenue by Category', fontsize=14, fontweight='bold')
ax.set_xlabel('Category', fontsize=12)
ax.set_ylabel('Total Revenue ($)', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('category_Sales_bar_Chart.png')
plt.show()
```

```python
#Count Plot: Regions-Wise Transcation 
region_counts = df['Region'].value_counts().reset_index()
region_counts.columns = ['Region', 'Count']
region_counts = region_counts.sort_values(by='Count', ascending=False)
fig,ax = plt.subplots(figsize=(8,6))
sns.barplot(data=region_counts,x='Region',y='Count', ax=ax, palette='viridis')
ax.set_title('Transaction distribution by Region', fontsize=14, fontweight='bold')
ax.set_xlabel('Region', fontsize=12)
ax.set_ylabel('Number of Transactions', fontsize=12)
plt.tight_layout()
plt.savefig('Region_transaction_count_plot.png')
plt.show()
```

**6. Summarize key findings**

**1. Overall Business Health (KPIs)**

The analysis of the sales database over the tracking period reveals a healthy revenue-to-profit conversion rate.

- **Total Revenue Generated:** $77,906.49

- **Total Profit Realized:** $22,495.46

- **Average Order Value (AOV):** $389.53

**Strategic Insight:** An Average Order Value of close to $390 indicates that transactions generally consist of high-value items or multi-product bundles, suggesting strong baseline purchasing power from the customer segments.

**2. Category Performance Analysis**

Segmenting the data by product categories highlights where the core financial volume is concentrated:

- **Home & Furniture:** This category emerged as the clear market leader, securing $29,550.49 in revenue and the highest profit contribution at $8,349.13. It also led operational volume with 72 total transactions.

- **Accessories:** A highly competitive second place, pulling in $25,545.78 in revenue. It maintains a very efficient profit-to-revenue ratio, proving to be a highly viable category for sustained investment.

- **Clothing & Apparel:** While ranking third, it still maintains a strong footprint with $22,810.22 in revenue and 65 transactions, showing a stable, consistent market demand.

**3. Geographical and Trend Insights**

- **Regional Dominance:** The West Region represents the highest transaction volume with 57 completed orders, followed closely by the South and Centre regions (51 orders each). The East Region showed the lowest engagement with 41 transactions.

- **Peak Commercial Period:** Chronological trend tracking identified May 2024 as the peak sales month, generating $6,340.13 in revenue. This spikes above the baseline monthly average, indicating powerful seasonal purchasing behavior or highly successful mid-year promotional campaigns.

**4. Data-Driven Recommendations for the Organization**

Based on the empirical insights extracted during this Exploratory Data Analysis (EDA), the following business actions are recommended:

- **Inventory Optimization:** Allocate a higher percentage of working capital and inventory capacity to the Home & Furniture and Accessories segments, as they drive both top-line revenue and bottom-line profit.

- **Targeted Regional Marketing:** Design localized marketing campaigns specifically targeting the East Region to address the lower transaction counts and bridge the geographic performance gap.

- **Seasonal Planning:** Replicate the promotional, pricing, or supply chain strategy utilized in May across other slower quarters of the fiscal year to stabilize monthly revenue fluctuations.
